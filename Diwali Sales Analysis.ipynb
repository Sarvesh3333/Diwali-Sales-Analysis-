## EDA

```python
sns.countplot(x = 'Gender', data =df)
```

```python
ax = sns.countplot(x = 'Gender', data =df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y='Amount', data = sales_gen)
```

*from above graphs we can see that most of the buyers are female and even the purchasing power of female is greater that men*

## Age Group

```python
df.columns
```

```python
sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
```

```python
sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data = sales_age)
```

*From above graph we can see that most of the buyers of age between 26-35 years and they are female*

```python
df.columns
```

```python
# Total number of orders from top 10 States

sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Orders', data = sales_state)
```

```python
#Total Amount/Sales from top 10 State

sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Amount', data = sales_state)
```

*From above graphs we can see that total number of orders, AMount/Sales are from UP, Maharashtra & Karnataka Respectively*


## Marital_status

```python
ax=sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(10,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#from above graph we can see that most of the buyers are married(women) and they have high purchasing power
```

## Occupation

```python
df.columns
```

```python
ax=sns.countplot(data = df, x = 'Occupation')

sns.set(rc={'figure.figsize':(20,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y='Amount', data = sales_occupation)
```

*From above graph we can see that most of the buyers are from IT Sector, Healtcare and Aviation sector*

## Product Category

```python
df.columns
```

```python
ax=sns.countplot(data = df, x = 'Product_Category')

sns.set(rc={'figure.figsize':(20,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category', y='Amount', data = sales_product)
```

*from above graph we can see that most sold products are from Food, Clothing&Apparel and Electronics&Gadgets*
